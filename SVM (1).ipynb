```python
import pandas as pd
import numpy as np
import optuna
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Load dataset (Wine Quality - red wine from UCI)
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'
df = pd.read_csv(url, sep=';')

# ----------------- Basic Data Analytics -----------------
print("Dataset Info:")
print(df.info())
print("\nMissing Values:\n", df.isnull().sum())
print("\nClass Distribution:\n", df['quality'].value_counts())

# ----------------- Feature-Label Split -----------------
X = df.drop('quality', axis=1)
y = df['quality']
scaler = StandardScaler()
X = scaler.fit_transform(X)

# ----------------- Optimization Setup -----------------
results = []
best_sample = {'index': None, 'accuracy': 0, 'study': None}

# Perform for 10 random samples
for sample_id in range(1, 11):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=sample_id)

    def objective(trial):
        kernel = trial.suggest_categorical('kernel', ['rbf', 'poly', 'sigmoid'])
        C = trial.suggest_float('C', 0.1, 100.0, log=True)
        gamma = trial.suggest_categorical('gamma', ['scale', 'auto'])

        model = SVC(kernel=kernel, C=C, gamma=gamma)
        model.fit(X_train, y_train)
        preds = model.predict(X_test)
        acc = accuracy_score(y_test, preds)
        return acc

    study = optuna.create_study(direction='maximize')
    study.optimize(objective, n_trials=100)

    best_acc = study.best_value
    best_params = study.best_params

    print(f"✅ Sample S{sample_id}: Accuracy = {best_acc:.4f}, Params = {best_params}")

    results.append({
        'Sample': f"S{sample_id}",
        'Accuracy': best_acc,
        'Parameters': best_params,
        'Study': study
    })

    if best_acc > best_sample['accuracy']:
        best_sample = {'index': sample_id, 'accuracy': best_acc, 'study': study}

# ----------------- Table 1 Output -----------------
table = pd.DataFrame({
    'Sample #': [r['Sample'] for r in results],
    'Best Accuracy': [round(r['Accuracy'] * 100, 2) for r in results],
    'Best SVM Parameters (Kernel, C, Gamma)': [
        f"{r['Parameters']['kernel']}, {r['Parameters']['C']:.2f}, {r['Parameters']['gamma']}" for r in results
    ]
})
print("\n📋 Table 1: Comparative performance of Optimized-SVM")
print(table.to_string(index=False))

# Optional: save table to CSV
table.to_csv("svm_results_svc.csv", index=False)

# ----------------- Plot Convergence of Best Sample -----------------
trial_values = [trial.value for trial in best_sample['study'].trials if trial.value is not None]

plt.figure(figsize=(8, 5))
plt.plot(trial_values, marker='o', linewidth=2, color='teal')
plt.xlabel("Iteration")
plt.ylabel("Accuracy")
plt.title(f"Figure 1: Convergence graph of best SVM (Sample S{best_sample['index']})")
plt.grid(True)
plt.tight_layout()
plt.savefig("best_svm_convergence_svc.png")
plt.show()
```
